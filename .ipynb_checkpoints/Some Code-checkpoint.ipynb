```python
from bs4 import BeautifulSoup
import requests
import re

def analyze_page(url):
    # Gets html from url and filters out the tags so that you have just the text, which it runs sentiment analysis on.
    html = BeautifulSoup(requests.get(url).text, 'lxml')
    pickuplines = []
    s = ''
    for p in html.find_all('li'):
        s = re.sub(r'<.+?>'," ", str(p).lstrip())
        sp = s.split("\n")
        pickuplines.append(sp)
    return pickuplines

h = analyze_page('http://pun.me/pages/funny-compliments.php')
```

```python
f = []
for i in h:
    for x in i:
        if x == '':
            i.remove(x)
        else:
            s = (' '.join(x.split()))
            f.append(s)
        

f
```
